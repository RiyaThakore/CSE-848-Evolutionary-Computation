Types of selection mentioned include:

### A)  Mating Selection (choosing who will become "parents"):

<!-- -->

1)  **Fitness-proportional selection (or "proportional selection"), in
    which individuals are chosen for mutation or crossover or continuing
    unmodified into the next generation based on their relative
    fitnesses (relative to others in the current population)**

    a.  Original GAs used this form

    b.  Disadvantage: as population fitnesses rise, relative fitnesses
        approach 1, and little distinction is made between best and
        worst individuals. Therefore, it is rarely used today

    c.  Two major forms are used for fitness-proportional selection:

        i.  Roulette Wheel Selection, in which a portion of the
            probability space [0., 1.] is allocated to each
            individual, proportional to its relative fitness, and
            uniformly chosen random numbers determine each choice of an
            individual to become a parent

        ii. Stochastic Universal Sampling, which uses the relative
            fitness distribution as above, but choice of a single random
            number (a "bias") determines the initial individual chosen,
            then 1/popsize selectors, evenly spaced, are added to that
            bias to determine all of the remaining parents---whatever
            "pie slice" a selector hits is sampled. This produces a
            fitness-proportional distribution with lower variance than
            Roulette Wheel Selection.

2)  Scaled fitness-proportional selection, in which the ratio of scaled
    fitness values of the best-fitness and mean-fitness individuals is
    kept constant (or some similar scheme)

    d.  Can help to avoid the problem in 1) b. above, but then the
        fitness of a chromosome varies with each generation, which may
        sometimes be a disadvantage

3)  Rank-based selection, in which individuals are sorted by fitness
    values, then each rank has its own specified chance of being chosen

    e.  Requires sorting of population at each generation, which could
        be costly for large population sizes

    f.  Still leaves question of what fraction should be allocated to
        each rank for the user to decide

4)  **Tournament selection, in which, for each parent to be chosen,
    *tourneysize* individuals are picked at uniform random, and the
    highest-fitness individual of those *tourneysize* individuals is
    chosen to be a parent**

    g.  Requires only one additional parameter, the tourneysize

    h.  Selection pressure can be controlled in a relatively constant
        manner by a single choice of tourneysize

    i.  Not subject to requirements to sort or rescale, nor does ability
        to distinguish among fitnesses shrink as evolution progresses

    j.  Very commonly used in practice

5)  Variations on tournament selection in which each individual is
    guaranteed a spot in one tournament, producing similar results but
    with lower variance in probability of selection

    k.  As above, but allows reduction in the variance among parents
        selected

6)  Variations on tournament selection, such as picking the top two of
    *tourneysize* individual as the ones to be crossed over

    l.  Can reduce computational overhead, but usually only to a trivial
        extent

    m.  Slightly weakens selection pressure compared with 4) or 5) above

7)  **Uniform probability of being selected for survival, mutation,
    crossover... i.e., all selection pressure is applied at the survival
    stage, not in selecting individuals to be mutated, crossed over, or
    continuing unmodified**

    n.  Nature of selection and selection pressure then depend only on
        the selection at the survival stage

8)  Mating restrictions exert a form of selection pressure, in a less
    direct way

    o.  May be useful if the nature of the problem makes certain types
        of matings undesirable, infeasible, etc.

    p.  By prohibiting certain matings, affects the probability
        distribution of matings among the potential parents, so affects
        selection pressure, but in a way that is difficult to quantify

### B)  Survival/Replacement Selection (of individuals to survive into next
    generation after mutation/crossover performed)

<!-- -->

1)  **Offspring replace their parents deterministically**

    a.  Original replacement strategy used in simple GAs

    b.  Can help to slow convergence, but also allows possible loss of
        the best individual found to date, unless combined with some
        form of elitism

2)  **Offspring replace their parents if their fitness is better (than
    one or than both)**

    c.  Stronger selection pressure than 1), which might cause more
        rapid convergence than desired (or might not, of course)...

    d.  Less selection pressure than Mu + Lambda (below), as the
        selection is more "local"

3)  **Offspring generated replace any parents (not necessarily their
    own), usually those parents with below-average or the worst
    fitnesses**

    e.  Replacing worst in population is frequently used, and doesn't
        seem to introduce high selection pressure

4)  **Offspring generated are pooled with all parents and popsize of
    them survive (similar to (Mu + Lambda) below, but with no
    specification of size of Mu relative to Lambda**

    f.  Preserves elite parents and children

    g.  Similar to 6) below, and commonly used, but can cause fairly
        rapid convergence (depending on the problem)

5)  **(Mu, Lambda) selection, in which the Lambda parents generate
    Mu\>Lambda offspring, and the best Lambda of those offspring replace
    all Lambda parents**

    h.  Commonly used with Evolutionary Strategies operators, but
        increasingly, also with other evolutionary algorithms

    i.  Does not have elite preservation property of 2), 4) and 6)

    j.  Less selection pressure than 6; sometimes 5) works better;
        sometimes 6)

6)  **(Mu + Lambda) selection, in which the Lambda parents generate
    Mu\>Lambda offspring, and the best Lambda of the (Mu + Lambda)
    parents and offspring survive**

    k.  Commonly used with Evolutionary Strategies operators, but
        increasingly, also with other evolutionary algorithms

    l.  Does preserve elite solutions (i.e., best solution found is
        never lost)

    m.  Traded off with 5); each is sometimes better

# Task 2 - Question

Let S1={1\*\*\*10\*\*00\*\*} and S2={\*\*\*1\*01\*\*\*\*\*} be two
schemata.

1\. Give the order and the defining length of S1 and S2.

2\. What is the probability for one-point crossover with crossover rate
$p_c$ that crossover breaks S1, or S2? (i.e., the probability that the
child created by the operator does not belong to the given schema,
assuming each is crossed over with a mate that does not match it in any
of the positions with a value specified.)

3\. What is the probability that mutation with mutation rate $p_m$ breaks
S1, or S2?

4\. What is the probability that S1 or S2 survive the application of both
crossover and mutation?

5\. Is it correct to call one of these two schemata a "building block"?
Explain why, or why not.


# Task 2 - Answer

1.  Order (the number of positions specified, not \*'s) of S1, o(S1) =
    5; o(S2) = 3. Defining length ∆ (number of spots 1-point crossover
    could disrupt) of S1, ∆(S1) = 9; ∆(S2) = 3.

2.  Assuming that crossover is with an individual that does not match S1
    at any locus (extremely unlikely for binary chromosomes, but much
    more likely otherwise), the probability that 1-point crossover
    disrupts S1 is ∆(S1) / (len (S1)-1) = 9/11 ~= 0.8181.

    Similarly, the probability that 1-point crossover disrupts S2 is
    ∆(S2) / (len (S2)-1) = 3/11 = 0.2727.

3.  Probability that mutation (at rate $p_m$ per locus) does not change a
    particular locus is (1-$p_m$). Therefore mutation does NOT break S~1~
    with probability (1 - $p_m$) \*\* o(S1) = (1 - $p_m$)\*\*5; that is,
    in order NOT to disrupt the schema requires that all o(S1) loci NOT
    be disrupted, so is the product of their probabilities. Therefore,
    the probability that mutation DOES break S1 is 1 - (1 - $p_m$)\*\*5.
    Similarly, mutation breaks S2 with probability 1 - (1 - $p_m$)\*\*o(S2) =

    1 - (1 - $p_m$)\*\*3. That is, NOT to disrupt S requires NOT
    disrupting at each of the o(S) positions, each of which occurs with
    probability (1 - $p_m$), and there are o(S) of those ALL of which must
    not be disrupted, so probabilities multiply. The chance of
    disrupting the schema is one minus the chance of NOT disrupting it.

4.  To survive both crossover and mutation, a schema must NOT be
    disrupted by mutation AND not disrupted by crossover, so we multiply
    those NOT disrupted probabilities to get the probability that
    neither disrupts the schema. Probability that schema S is not
    disrupted by *crossover* is (1 - ∆(S) / (len (S)-1)). Probability that
    schema S is not disrupted by *mutation* is (1 - pm)\*\*o(S) . So
    for schema S1, probability **not disrupted** by either crossover or
    mutation is:

    (1 - (∆(S1) / (len (S1) - 1)) \* (1 - $p_m$)\*\*o(S1) = (1 - .8181) \* (1 -
    $p_m$)\*\*5 ~= .1818 \* (1 - $p_m$)\*\*5

    And for schema S2:

    (1 - (∆(S2) / (len (S2) - 1))) \* (1 - $p_m$)\*\*o(S2) = (1 - .2727) \* (1 -
    $p_m$)\*\*3 ~= .7272 \* (1 - $p_m$)\*\*3

5.  A building block is defined as a schema that is:

    a\) short,

    b\) low-order, and

    c\) high-fitness

    S1 is NOT a building block. Relative to a total chromosome length
    of 12 bits, 9 (= defining length of S1) seems long, so S1 would not
    be considered a building block (it's also not very low-order, its
    order being 5).

    The defining length of S2, namely 3, is much shorter relative to 12,
    and its order is only 3, which is relatively low-order, so IF S2 is
    of relatively **high fitness**, it might well be considered a
    building block. If its fitness is not high, then it would NOT be
    considered a building block.


# Task 3 - Solution and Discussion 

The problem asks for a comparison of a GA versus enumeration for finding a solution to the Eight Queens problem.  There is something different about this problem, however, from the usual enumeration problem:  in the “usual” enumeration problem, one must calculate and evaluate all possible solutions (i.e., enumerate them all) in order to know which one is best.  But in the Eight Queens problem, there is a solution criterion known in advance (namely, 0 attacks) that allows the enumeration to be stopped as soon as any solution that satisfies it is found.  Similarly, we can (and I hope you did!) stop the GA whenever it finds a solution with no attacks.  It would be a hard task for a GA to find ALL solutions to this problem—we can stop it as soon as it finds one solution with no attacks.  Thus, to compare the two algorithms, what we should be comparing is the average number of evaluations it takes a GA to find ONE solution versus the average number of evaluations it takes enumeration to find ONE solution (its first legal solution).  The easiest way to do that enumeration task, in this small case with only 8 queens, is to run the enumeration once, tabulating the number of permutations and the number of them that are legal solutions.  Then one can calculate the average number of evaluations needed to find a solution via enumeration, assuming it were started from a random initial arrangement of queens (with only one per row and column).  

### Exhaustive Search

The ’C’ program appended does that, finding that there are 40,320 total permutations of the queens, and 92 of them are legal solutions to the eight queens problem.  That allows calculating the average time to find a solution by enumeration (if done from random starting points) as 443.26 evaluations.  That is about double the number of evaluations required by the GA (see Table I below), which is about 220, and, of course, that would become far more dramatic if the number of queens were larger.  I also implemented a version of the exhaustive search that starts from a random arrangement (one queen per row/column) and permutes them by Heap’s Algorithm until finding a first correct solution, then stops. The results of that algorithm are shown in the second column of Table I, and compared statistically with the GA results.

### Discussion

There are many ways to attack the enumeration problem, including:

1) You might enumerate all possible combinations of 8 queens among 8 the columns, which is 8**8, or 16,777,216 arrangements.  But that ignores the fact that we are not allowing any row to contain more than one queen, so is not a good way to solve this problem.

2) You might recognize that you could enumerate column by column, so have 8 choices for the first column, then 7 for the second, then 6 for the third, etc., yielding 8! or 40,320 possibilities in total to be checked for any attacks. 

3) You might recognize that each of these 8! solutions that has only one queen in any row or column is a permutation of any other such solution, a permutation of the numbers (1,2,3,4,5,6,7,8). (There are lots of ways to see this… including just permuting the COLUMNS of any initial solution (1 queen in each row, each column) without moving the rows of the queens in those columns, and there are 8! ways to do this. Each such permutation, of course, results in having only one queen in each row and each column.)  Storing the row index of the queen in each column is the “representation” that we suggested you use in your genetic algorithm solution of the problem.  Permuting these solutions is a much more efficient way to enumerate all possible board arrangements with exactly one queen per row or column.  There are only 8! or 40,320 such arrangements, so a much smaller number of arrangements to enumerate than 8**8.  We’ll discuss this in 4) below.

4) To attack the enumeration problem DIRECTLY, we need our enumeration algorithm to be capable of producing all possible permutations, each exactly once, starting from an arbitrary permutation, and that is non-trivial to do.  B. R. Heap invented such an algorithm in 1963.  I have implemented his algorithm in a C program (my native language) and run 30 instances of it, each beginning with a different randomly chosen permutation.  Since you can’t run it in the Jupyter Notebook environment, I haven’t listed it here, but instead listed the code for the solution enumerating all solutions.  

5) Of course, you could use the method I did above under “Exhaustive Search” and enumerate ALL solutions and divide the size of the search space by the total number of solutions found.  Perhaps some of you have done that.  


But I also implemented the direct enumeration approach, stopping as soon as a legal solution is found. The difference in means between the GA and enumeration approaches appears obvious, but perhaps not as large as a naïve comparison of average GA run time with TOTAL time to enumerate ALL solutions might have led you to believe—there are many solutions (namely, 92) that satisfy the 0-attacks criterion, so the average or median time to find the first one of them is much less than the total time to enumerate all solutions.  Here, 40,320/92 = 438.26, which is near the mean of what our direct (stopping at first success) enumeration algorithm below produced.

If we wanted to use these runs to demonstrate that the mean solution times of the two algorithms are different statistically, we would perform a statistical test on data such as those in Table I.  Since we have no assurance that these evaluation numbers are normally distributed, rather than using something like a Student t-test, we use a Mann-Whitney U Test (also called the Wilcoxon Rank Sum Test or the Mann-Whitney-Wilcoxon Test), which sorts all 60 values, and then calculates the probability that the resulting order of outputs from the two algorithms would occur in that fashion due to chance.  If the samples were paired rather than independent, we would instead use the Wilcoxon Signed Rank Test, a very similar test also heavily used in analyzing data produced by evolutionary computation.  Doing this Mann-Whitney U Test on these data produces the result that the difference is statistically significant at the level of 0.01044.  That is, the chance that such a set of results would occur if the two underlying distributions were identical is less than 0.01044, well below the level of .05 often used for such testing, but not so ridiculously small as to be essentially impossible.   This is a typical test used to compare results generated by evolutionary computation algorithms, if there is no assurance that their distributions are Gaussian (normal).

    ## ‘C’ Code for Generating ALL Legal Solutions to the Eight Queens Problem via Enumeration (WILL NOT EXECUTE IN THIS CONTEXT... COPY AND USE gcc <filename> to compile, use a.out to execute.

```python
/* Routine to count the solutions to 8 queens problem by exhaustive search */
/* This version calculates ALL legal solutions by exhaustion. */
#include <math.h>
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define NQUEENS 8

int  CheckAttacks(int *Queens);
void generate(int k, int *Queens);
void swap(int *first, int *second);

int  NPermsChecked = 0;
int  NSolutions = 0;

int
main(void)
{
   int Queens[NQUEENS] = {0,1,2,3,4,5,6,7};
   time_t t;
   int seed;

   seed = time(NULL);
   srand(seed);
   /* Just use 0-7 in order as starting arrangement; will enumerate all, */
   /* so it doesn't matter where start. */
   /* Generate all possible permutations of Queens, starting from this */
   /* starting arrangement, using Heap's Algorithm (see Wikipedia entry).   */
   /* At each step, the current permutation is stored in Queens. */
   /* Calculate all solutions, counting both number tried and number found. */
   generate(NQUEENS, Queens);
   fprintf(stdout, "\nPermutations %d, Solns Found %d\n", NPermsChecked, NSolutions);
}

void
generate(int k, int *Queens)
{
   /* Implementation of Heap's Algorithm to enumerate permutations */
   int i, row, Attacks;

   if (k == 1) {
      NPermsChecked++;
      /* Permutation complete; check this permutation for legal, record successes. */
      NSolutions += !CheckAttacks(Queens);
   }
   else {
      for (i=0;i<k;i++) {
          generate(k - 1, Queens);
          if (k % 2 == 0) swap(&(Queens[0]), &(Queens[k-1]));
          else swap(&(Queens[i]), &(Queens[k-1]));
      }
   }
   return;
}

int
CheckAttacks(int *Queens)
{
   /* Check queen in each row to see if it attacks a queen in later rows. */
   /* Can stop when find first attack, of course. */
   int row, row2, CkQueen;
   int Attacks = 0;

   for (row=0;row<NQUEENS;row++) {
      CkQueen = Queens[row]; /* column of queen in row row. */
      for (row2=row;row2<NQUEENS;row2++) {
         if (row != row2 && abs(row-row2) == abs(CkQueen-Queens[row2])) {
            /* equal absolute values means on same diagonal */
            Attacks = 1;
            return (Attacks);
         }
      }
   }
   return (Attacks);
}

void
swap(int *first, int *second)
{
   int temp;

   temp = *first;
   *first = *second;
   *second = temp;
   return;
}
```

#### Step 1: Make a scoring function

```python
def objfn(geno):
    """
    8 queens objective function.
    max = 28
    min = 0
    Goal is to minimize.
    
    Parameters
    ----------
    geno : numpy.ndarray
        1d array representing the genotype of an individual.
    
    Returns
    -------
    violations : int
        Number of violations for queens
    """
    violations = 0
    for i in range(len(geno)):
        for j in range(i+1, len(geno)):
            # calculate slope
            slope = (geno[j]-geno[i]) / (j-i)
            # check row and diagonal violations
            if (geno[i] == geno[j]) or (slope == 1) or (slope == -1):
                violations += 1
    return violations
```

#### Step 2: Make an exhaustive search algorithm to search for the chess arrangement

```python
def permute(a, b):
    """
    A recursive generator function to yield permuations of an input.
    
    Parameters
    ----------
    a : numpy.ndarray
        Input array to be sampled from.
    b : numpy.ndarray
        Output array to be yielded if the lenght of a is 0.
    
    Yields
    ------
    b : numpy.ndarray
        The output array if len(a == 0)
    """
    if len(a) == 0:
        yield numpy.array(b)
    for i in range(len(a)):
        btmp = numpy.append(b, a[i])
        atmp = numpy.delete(a, i)
        yield from permute(atmp, btmp)
```

```python
def exhaustive_search(geno, target = 0):
    """
    Perform exhaustive search for 8 queens problem.
    
    Parameters
    ----------
    geno : numpy.ndarray
        Starting board configuration to work with.
    target : int
        Target score
    
    Returns
    -------
    k : numpy.ndarray, None
        Found board configuration or None if no boards are found.
    ntests : int
        Number of function evaluations
    """
    out = numpy.array([], dtype=geno.dtype)
    ntests = 0
    for k in permute(geno, out):
        score = objfn(k)
        ntests += 1
        if score == target:
            return k, ntests
    return None, ntests
```

##### Create initialization function

```python
def initialize(n, p):
    """
    Initialize breeding population
    
    Parameters
    ----------
    n : int
        Number of breeding individuals
    p : int
        Number of rows/columns on the chess board.
    
    Returns
    -------
    pop : numpy.ndarray
        A 2D array of shape (n, p) containing breeding individuals
    """
    # copy 1-p, n times
    pop = numpy.tile(numpy.arange(p)+1, (n,1))
    
    # shuffle along the second axis
    for i in range(n):
        numpy.random.shuffle(pop[i])
        
    return pop
```

##### Create population evaluation function

```python
def evaluate(pop, fn, sign = 1):
    """
    Evaluate a population using an objective function
    
    Parameters
    ----------
    pop : numpy.ndarray
        A 2D array containing breeding individual chromosomes
    fn : callable
        Objective function
    sign : int
        1 or -1 to alter objective function sign
        
    Returns
    -------
    score : numpy.ndarray
        A 1D array containing scores for the population.
    """
    score = numpy.array(
        [(sign * fn(pop[i])) for i in range(len(pop))]
    )
    return score
```

##### Create parental selection operator: uniform selection

```python
def uniform_parental_selection(pop, pscore, mu):
    """
    Perform uniform parental selection selection
    
    Parameters
    ----------
    pop : numpy.ndarray
        A 2D array containing breeding individual chromosomes
    pscore : numpy.ndarray
        A 1D array containing scores for the population.
    mu : int
        Number of parents to select
    
    Returns
    -------
    parents: numpy.ndarray
        A 2D array containing parental chromosomes.
    score : numpy.ndarray
        A 1D array containing scores for the selected parents.
    """
    parents = numpy.random.choice(
        numpy.arange(len(pop)),
        size=mu,
        replace=True
    )
    return pop[parents], pscore[parents]
```

##### Create recombination operator: Recombines using student's own, looks like single-point crossover with repair to work with permutation representation, with probability p.

```python
def recombine_1pt(male, female, p = 1.0):
    """
    Recombine (single-point) male and female individuals to produce 2 progeny per pair.
    
    Parameters
    ----------
    male : numpy.ndarray
        A 2D array containing male parental chromosomes.
    female : numpy.ndarray
        A 2D array containing female parental chromosomes.
    p : int
        Probaility of recombination for a given pair of individuals.
    
    Returns
    -------
    children : numpy.ndarray
        A 2D array containing progeny chromosomes.
    """
    children = []
    for a, b in zip(male,female):
        atmp, btmp = a, b

        # determine if we're going to crossover
        xo = numpy.random.choice([True,False], p=[p,1-p])
        if xo:
            # copy arrays
            atmp, btmp = numpy.copy(a), numpy.copy(b)

            # crossover occurs directly before this index [1,7]
            xopos = numpy.random.randint(1, len(a))

            # crossover a with b
            i = xopos
            for e in numpy.append(b[xopos:],b[:xopos]):
                if not (e in atmp[:xopos]):
                    atmp[i] = e
                    i += 1
            # crossover b with a
            i = xopos
            for e in numpy.append(a[xopos:],a[:xopos]):
                if not (e in btmp[:xopos]):
                    btmp[i] = e
                    i += 1
        
        children.append(numpy.copy(atmp))
        children.append(numpy.copy(btmp))
    
    return numpy.stack(children)
```

##### Create the mutation operator: Randomly exchanges loci with probability p per locus.

```python
def mutate(pop, p):
    """
    Mutate loci by exchanging values at different loci along the same chromosome for an individual.
    
    Parameters
    ----------
    pop : numpy.ndarray
        A 2D array containing population chromosomes.
    p : int
        Probaility of mutation for a single locus.
    
    Returns
    -------
    children : numpy.ndarray
        A 2D array containing progeny chromosomes.
    """
    children = []
    for a in pop:
        # generate mask of loci to mutate
        mut = numpy.random.choice([True,False], size=a.shape, replace=True, p=[p,1-p])

        # loop through the mask to change values
        for i,m in enumerate(mut):
            if m == True:
                # generate index to exchange
                j = i
                while (j == i):
                    j = numpy.random.randint(0,len(a))
                # exchange indices
                a[i], a[j] = a[j], a[i]
        children.append(numpy.copy(a))
    
    return numpy.stack(children)
```

##### Create child selection operator: mu + lambda selection

```python
def parent_child_selection(parents, pscore, children, cscore, mu):
    """
    Select top individuals from both parents and progeny by lowest score.

    Parameters
    ----------
    parents : numpy.ndarray
        A 2D array containing parental chromosomes.
    pscore : numpy.ndarray
        A 1D array containing scores for the parental population.
    children : numpy.ndarray
        A 2D array containing progeny chromosomes.
    cscore : numpy.ndarray
        A 1D array containing scores for the progeny population.
    mu : int
        Number of parents to select
    
    Returns
    -------
    children : numpy.ndarray
        A 2D array containing progeny chromosomes.
    """
    # merge arrays
    pop = numpy.concatenate([parents,children])
    score = numpy.concatenate([pscore,cscore])
    
    # get top (lowest) scoring indicies
    ix = score.argsort()[:mu]
    
    return numpy.copy(pop[ix]), numpy.copy(score[ix])
```

##### Put everything together into evolutionary algorithm

```python
def queens_evolutionary_algorithm(mu, lmbda, fn, pXO, pMut, target = 0, nqueens = 8):
    """
    Evolve a population to solve the 8-queens problem.
    
    Parameters
    ----------
    mu : int
        Number of parents
    lmbda : int
        Number of progeny (must be even)
    fn : callable
        Objective function to optimize
    pXO : float
        Probability of a 1pt crossover
    pMut : float
        Probability of a mutation (per locus)
    target : numeric
        Target objective function value (0 for function minimum)
    nqueens : int
        Number of queens/rows/columns on the board
        
    Returns
    -------
    config : numpy.ndarray
        Configuration of the solution board
    score : numeric
        Score associated with 'config'.
    neval : int
        Number of function evaluations to optimize.
    """
    # track number of evaluations
    neval = 0
    
    # initialize population
    pop = initialize(mu, nqueens)
    oscore = evaluate(pop, fn)
    
    # add number of evaluations
    neval += len(oscore)
    
    while oscore.min() != target:
        # choose parents for lmbda progeny
        parents, pscore = uniform_parental_selection(pop, oscore, lmbda)
        
        # recombine and mutate to make children
        children = recombine_1pt(parents[0::2], parents[1::2], pXO)
        children = mutate(children, pMut)
        
        # evaluate new progeny
        cscore = evaluate(children, fn)
        
        # add number of evaluations
        neval += len(cscore)
        
        # select top individuals from the parents + children
        pop, oscore = parent_child_selection(parents, pscore, children, cscore, mu)
    
    # get index of best scorer
    tix = numpy.argwhere(oscore == target)[0][0]
    
    return pop[tix], oscore[tix], neval
```
