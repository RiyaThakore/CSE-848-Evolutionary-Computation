## 1-1. Monte Carlo Approach

```python
def monte_carlo_guesser_wrapper(desired):  
  initial = "".join([chr(n) for n in np.random.randint(ord('A'), ord('Z')+1, len(desired))])
  iterations = 0
  while initial != desired :
    initial = "".join([chr(n) for n in np.random.randint(ord('A'), ord('Z')+1, len(desired))])
    iterations += 1
  return iterations

def monte_carlo_guesser(desired, guesses=10, verbose=True):
  avg = sum([monte_carlo_guesser_wrapper(desired) for i in range(guesses)]) / guesses
  if verbose:
    print(f"[Monte Carlo] Average iterations among {guesses} guesses: {avg}")
  return avg

monte_carlo_guesser("COIN")
```

```python
iter_hist_MC = [monte_carlo_guesser("COIN"[:i], 10, False) for i in range(1, 5)]
```

```python
# Plot trends base on letter length
plt.figure(figsize=(12, 8))
plt.plot([i for i in range(1, 5)], iter_hist_MC)
plt.title("Average Iterations of Monte Carlo")
plt.xlabel("num letters")
plt.ylabel("average iterations")
plt.legend()
plt.plot()
```

## 1-2. Informed Search Approach

```python
def informed_search_wrapper(desired):  
  initial = "".join([chr(n) for n in np.random.randint(ord('A'), ord('Z')+1, len(desired))])
  iterations = 0
  while initial != desired:
    desired = "".join([desired[i] for i in range(len(desired)) if desired[i] != initial[i]])
    initial = "".join([chr(n) for n in np.random.randint(ord('A'), ord('Z')+1, len(desired))])
    iterations += 1
  return iterations

def informed_search_guesser(desired, guesses=10, verbose=True):
  avg = sum([informed_search_wrapper(desired) for i in range(guesses)]) / guesses
  if verbose:
    print(f"[Informed Search] Average iterations among {guesses} guesses: {avg}")
  return avg

informed_search_guesser("COIN", 10)
```

```python
iter_hist_IS = [informed_search_guesser("COINSAREGOOD"[:i], 1000, False) for i in range(1, 13)]
```

```python
# Plot trends base on letter length
plt.figure(figsize=(12, 8))
plt.plot([i for i in range(1, 13)], iter_hist_IS)
plt.title("Average Iterations of Informed Search")
plt.xlabel("num letters")
plt.ylabel("average iterations")
plt.legend()
plt.plot()
```

```python
# Combined Plot trends base on letter length
plt.figure(figsize=(12, 8))
plt.plot([i for i in range(1, 13)], iter_hist_IS, label="Informed Search")
plt.plot([i for i in range(1, 5)], iter_hist_MC, label="Monte Carlo")
plt.title("Average Iterations of Informed Search")
plt.xlabel("num letters")
plt.ylabel("average iterations")
plt.legend()
plt.plot()
```

**Conclusion:** Based on the plots, we can state that Monte Carlo's search is exponential and Informedd Search is polynomial

* **Output**: The classified traffic sign.

* **Input**: The time elapsed given from a starting time.
* **Output**: The location of each planet around the sun.
