## Monte-Carlo/Random Search

### Specific

```python
import random

first='C'
second='O'
third='I'
fourth='N'
org = first + second + third + fourth
new = ''
iterations = 0
while new!=org:
  a = chr(random.randint(ord('A'), ord('Z')))
  b = chr(random.randint(ord('A'), ord('Z')))
  c = chr(random.randint(ord('A'), ord('Z')))
  d = chr(random.randint(ord('A'), ord('Z')))
  new = a + b + c + d
  iterations+=1
print('Org:', org)
print('New:', new)
print('Iterations:', iterations)
```

```python
import random
def monte_carlo_search():
  first='C'
  second='O'
  third='I'
  fourth='N'
  org = first + second + third + fourth
  new = ''
  iterations = 0
  while new!=org:
    a = chr(random.randint(ord('A'), ord('Z')))
    #print(a)
    b = chr(random.randint(ord('A'), ord('Z')))
    #print(b)
    c = chr(random.randint(ord('A'), ord('Z')))
    #print(c)
    d = chr(random.randint(ord('A'), ord('Z')))
    #print('d',d)
    new = a + b + c + d
    iterations+=1
  return iterations

def monte_carlo_iterations():
  sum = 0
  print('Monte-Carlo Search')
  for i in range(10):
    print('******', i, ' iteration done *******')
    iter = monte_carlo_search()
    print(iter, ' number of iterations')
    sum += iter
  return sum/10

print('Average number of iterations:',monte_carlo_iterations())
```

### Generic

```python
def monte_carlo_general_search(desired_word):  
  initial = "".join([chr(i) for i in random.randint(ord('A'), ord('Z'), len(desired_word))])
  iterations = 0
  while initial != desired_word :
    initial = "".join([chr(i) for i in random.randint(ord('A'), ord('Z'), len(desired_word))])
    iterations += 1
  return iterations

def monte_carlo_final(desired_word):
  guesses = 10
  mc_sum = [monte_carlo_general_search(desired_word) for g in range(guesses)]
  mc_avg = mc_sum / guesses
  return mc_avg

monte_carlo_final("COIN")
```

## Directed Search/Informed Search

### Specifc

```python
import random
first='C'
second='O'
third='I'
fourth='N'
org = first + second + third + fourth
new = ''
iterations = 0
a = ''
b = ''
c = ''
d = ''
while new!=org:
  if a != first:
    a = chr(random.randint(ord('A'), ord('Z')))
  if b != second:
    b = chr(random.randint(ord('A'), ord('Z')))
  if c != third:
    c = chr(random.randint(ord('A'), ord('Z')))
  if d != fourth:
    d = chr(random.randint(ord('A'), ord('Z')))
  new = a + b + c + d
  iterations+=1
print('Org:', org)
print('New:', new)
print('Iterations:', iterations)
```

```python
import random
def directed_search():
  first='C'
  second='O'
  third='I'
  fourth='N'
  org = first + second + third + fourth
  new = ''
  iterations = 0
  a = ''
  b = ''
  c = ''
  d = ''
  while new!=org:
    if a != first:
      a = chr(random.randint(ord('A'), ord('Z')))
    if b != second:
      b = chr(random.randint(ord('A'), ord('Z')))
    if c != third:
      c = chr(random.randint(ord('A'), ord('Z')))
    if d != fourth:
      d = chr(random.randint(ord('A'), ord('Z')))
    new = a + b + c + d
    iterations+=1
  return iterations

def directed_search_iterations():
  sum = 0
  print('Directed Search')
  for i in range(10):
    print('******', i, 'th iteration done *******')
    iter = directed_search()
    print(iter, ' number of iterations')
    sum += iter
  return sum/10


print('Average number of iterations:',directed_search_iterations())
```

### Generic

```python
def informed_search_general(desired_word):  
  new = "".join([chr(i) for i in random.randint(ord('A'), ord('Z'), len(desired_word))])
  iterations = 0
  while new != desired_word:
    desired_word = "".join([desired_word[j] for j in range(len(desired_word)) if desired_word[j] != new[j]])
    new = "".join([(chr(k)) for k in random.randint(ord('A'), ord('Z'), len(desired_word))])
    iterations += 1
  return iterations

def informed_search_final(desired_word):
  guesses = 10
  is_sum = [informed_search_general(desired_word) for g in range(guesses)]
  is_avg = is_sum / guesses
  return is_avg

informed_search_final("COIN")
```
